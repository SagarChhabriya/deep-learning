```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras.preprocessing import  image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img(r"D:\Ao\Code\AI\deep-learning\notes\11-CNN\train\cat.jpeg", target_size=(250,250))
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,250,250,3)
```

```python
# flow: for single image
# flow_directory: for set of input images

i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'):
    i = i + 1
    # This will run infinite: I need only 10 images as examples
    if i == 10:
        break
```
