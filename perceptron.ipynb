```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = [
    {"id": 1, "x1": 8.2, "x2": 7.5, "y": 1},
    {"id": 2, "x1": 7.1, "x2": 6.0, "y": 1},
    {"id": 3, "x1": 5.5, "x2": 5.0, "y": 0},
    {"id": 4, "x1": 6.8, "x2": 7.0, "y": 1},
    {"id": 5, "x1": 4.0, "x2": 3.5, "y": 0},
    {"id": 6, "x1": 9.0, "x2": 8.5, "y": 1},
    {"id": 7, "x1": 5.0, "x2": 4.0, "y": 0},
    {"id": 8, "x1": 7.5, "x2": 6.5, "y": 1},
    {"id": 9, "x1": 6.0, "x2": 5.5, "y": 0},
    {"id": 10, "x1": 8.5, "x2": 9.0, "y": 1},
]

# Updated weights after first update
w1, w2, b = 0.45, -1.5, -0.1

# Visualization function
def plot_decision_boundary(data, w1, w2, b, step):
    sns.set(style="whitegrid")
    plt.figure(figsize=(8, 6))
    
    # Plot data points
    for point in data:
        color = "blue" if point["y"] == 1 else "red"
        plt.scatter(point["x1"], point["x2"], color=color, s=100)
    
    # Plot decision boundary
    x_values = np.linspace(3, 10, 100)
    y_values = -(w1 * x_values + b) / w2
    plt.plot(x_values, y_values, label=f"Decision Boundary (Step {step})", color="green")
    
    # Labels and legend
    plt.xlabel("CGPA (x1)")
    plt.ylabel("Resume Score (x2)")
    plt.title(f"Perceptron Training (Step {step})")
    plt.legend()
    plt.show()

# Visualize after first update
plot_decision_boundary(data, w1, w2, b, step=1)
```

```python
# Updated weights after second update
w1, w2, b = 1.27, -0.75, 0.0

# Visualize after second update
plot_decision_boundary(data, w1, w2, b, step=2)
```

```python
# Updated weights after third update
w1, w2, b = 1.95, 0.05, 0.1

# Visualize after third update
plot_decision_boundary(data, w1, w2, b, step=3)
```

```python
# Updated weights after fourth update
w1, w2, b = 1.55, -0.3, 0.0

# Visualize after fourth update
plot_decision_boundary(data, w1, w2, b, step=4)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create DataFrame
data = [
    {"x1": 8.2, "x2": 7.5, "y": 1}, {"x1": 7.1, "x2": 6.0, "y": 1},
    {"x1": 5.5, "x2": 5.0, "y": 0}, {"x1": 6.8, "x2": 7.0, "y": 1},
    {"x1": 4.0, "x2": 3.5, "y": 0}, {"x1": 9.0, "x2": 8.5, "y": 1},
    {"x1": 5.0, "x2": 4.0, "y": 0}, {"x1": 7.5, "x2": 6.5, "y": 1},
    {"x1": 6.0, "x2": 5.5, "y": 0}, {"x1": 8.5, "x2": 9.0, "y": 1}
]
df = pd.DataFrame(data)

# Current weights
w1, w2, b = 0.45, -1.5, -0.1

# Calculate decision boundary line
x_values = np.linspace(3, 10, 100)
y_values = -(w1 * x_values + b) / w2

# Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="x1", y="x2", hue="y", palette={1: "blue", 0: "red"})
plt.plot(x_values, y_values, label="Decision Boundary", color="green")
plt.xlabel("CGPA")
plt.ylabel("Resume Score")
plt.title("Perceptron Decision Boundary")
plt.legend()
plt.show()
```

```python
w1, w2, b = 1.27, -0.75, 0.0  # Updated weights
y_values = -(w1 * x_values + b) / w2

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="x1", y="x2", hue="y", palette={1: "blue", 0: "red"})
plt.plot(x_values, y_values, label="Decision Boundary (Update 2)", color="green")
plt.xlabel("CGPA"); plt.ylabel("Resume Score")
plt.legend(); plt.show()
```

```python
w1, w2, b = 0.45, -1.25, -0.2  # Converged weights
w1, w2, b = 1.55, -1.3, -0.1  # Correct final weights
y_values = -(w1 * x_values + b) / w2

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="x1", y="x2", hue="y", palette={1: "blue", 0: "red"})
plt.plot(x_values, y_values, label="Final Decision Boundary", color="green")
plt.xlabel("CGPA"); plt.ylabel("Resume Score")
plt.legend(); plt.show()
```
